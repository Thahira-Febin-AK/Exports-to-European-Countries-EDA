```python
df.isnull().sum()
```

```python
exchange_rate = 83   # Example: 1 USD = 83 INR

# Fill missing USD values
df['value_dl'] = df['value_dl'].fillna(df['value_rs'] / exchange_rate)

# Fill missing INR values
df['value_rs'] = df['value_rs'].fillna(df['value_dl'] * exchange_rate)
```

```python
df = df.dropna(subset=['value_rs', 'value_dl'])
```

```python
df['country_code'] = df['country_code'].fillna(0)
df['alpha_3_code'] = df['alpha_3_code'].fillna('UNKWN')
```

```python
df.fillna({
    'unit': 'Unknown',
    'value_qt': 'Unknown'
}, inplace=True)
```

```python
df = df.dropna()
```

```python
# Check missing values in each column
df.isnull().sum()
```

<h2 style="font-size: 20px; font-weight: bold; margin-bottom: 15px;">
 Treating outliers
</h2>

</div>

```python
cols = ['value_qt', 'value_rs', 'value_dl']

# Convert columns to numeric
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
cols = ['value_qt', 'value_rs', 'value_dl']

for col in cols:

    # Calculate IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    # Define lower and upper limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    print(f"\n{col}")
    print("Lower Limit :", lower_limit)
    print("Upper Limit :", upper_limit)

    # Cap outliers instead of removing rows
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)
```
